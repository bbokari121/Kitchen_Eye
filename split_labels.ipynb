```python
import os
import shutil

ORIGINAL_LABELS_DIR = "/home/sagemaker-user/Kitchen_Eye/data/Dataset/labels"
SPLIT_BASE_DIR = "/home/sagemaker-user/Kitchen_Eye/data/split_dataset"

splits = ["train", "val", "test"]

print("=== Start label splitting ... ===\n")

total_synced = 0
total_skipped = 0

for split in splits:
    img_dir = os.path.join(SPLIT_BASE_DIR, split, "images")
    lbl_dir = os.path.join(SPLIT_BASE_DIR, split, "labels")
    os.makedirs(lbl_dir, exist_ok=True)

    images = [f for f in os.listdir(img_dir)]
    images = [f for f in os.listdir(img_dir)]
    
    success_count = 0
    skip_count = 0
    missing_count = 0
    
    for img_name in images:
        base_name = os.path.splitext(img_name)[0]
        label_name = base_name + ".txt"
        
        src_label_path = os.path.join(ORIGINAL_LABELS_DIR, label_name)
        dst_label_path = os.path.join(lbl_dir, label_name)
        
        if os.path.exists(dst_label_path):
            total_skipped += 1
            continue

        if os.path.exists(src_label_path):
            shutil.copy2(src_label_path, dst_label_path)
            success_count += 1
            total_synced += 1
        else:
            missing_count += 1

    print(f"[{split.upper()}] Images: {len(images)} | Synced: {success_count} | Missing: {missing_count}")

print(f"\n=== Process Complete! ===")
print(f"Total Labels Synced:  {total_synced}")
print(f"Total Labels Skipped: {total_skipped}")
```
